```python
import glob
import cv2
import numpy as np
import onnxruntime as ort

# Load ONNX model
session = ort.InferenceSession("runs/segment/train14/weights/best.onnx")

# Get any test image
img_path = glob.glob("data/opg/test/images/*.jpg")[1]
print("Using:", img_path)

img = cv2.imread(img_path)
img_resized = cv2.resize(img, (512, 512))

img_input = img_resized.transpose(2, 0, 1)[None].astype(np.float32)

inputs = {session.get_inputs()[0].name: img_input}
output = session.run(None, inputs)

print("Inference OK!")

```
